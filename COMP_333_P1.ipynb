```python
# prompt: I want to remove duplicate entries in kingcoSales

# Remove duplicate entries based on 'sale_id' (assuming it's a unique identifier)
kingcoSales = kingcoSales.drop_duplicates(subset=['sale_id'], keep='first')
kingcoSales.describe()
```

```python
kingcoSales['sale_date'] = pd.to_datetime(kingcoSales['sale_date'])
kingcoSales['sale_year'] = kingcoSales['sale_date'].dt.year
print(kingcoSales['sale_year'].unique())
kingcoSales.describe()
```

```python
kingcoIndex['DATE'] = pd.to_datetime(kingcoIndex['DATE'])
kingcoIndex['year'] = kingcoIndex['DATE'].dt.year
kingcoIndex = kingcoIndex.drop('DATE', axis=1)  # Assign the result back to kingcoIndex
kingcoIndex = kingcoIndex.rename(columns={'ATNHPIUS53033A': 'HPI'})
print(kingcoIndex['year'].unique())
kingcoIndex.describe()
```

```python
# Merge the two DataFrames based on the year
merged_df = pd.merge(kingcoSales, kingcoIndex[['year', 'HPI']], left_on='sale_year', right_on='year', how='left')

# Calculate the inflation adjustment factor (using 2024 as the base year, you might need to change this)
base_year_hpi = merged_df.loc[merged_df['year'] == 2023, 'HPI'].iloc[0]
merged_df['inflation_factor'] = base_year_hpi / merged_df['HPI']

# Adjust the sale_price for inflation
merged_df['adjusted_sale_price'] = merged_df['sale_price'] * merged_df['inflation_factor']
merged_df = merged_df.drop('year', axis=1)
# You can now access the adjusted sale prices in the 'adjusted_sale_price' column
merged_df
```
